```python
# Insere documentos na nova base de dados
db.pontos_sp.insert_many(dicionario)
```

```python
# Estabelencendo um ponto central do mapa (Centro do Brasil)
center_lat = -14.24007
center_lon = -53.18050
```

```python
# Todos os dados  na variavel consulta
consulta = list(db.pontos_sp.find())
```

```python
# Converte em um data frame
df_convertido = pd.DataFrame(consulta)
```

```python
# Imprime a primeira linha do DF
df_convertido.iloc[0]
```

```python
# Criando um mapa a partir do ponto central com 800x600
locationmap = folium.Map(location=[center_lat , center_lon ], zoom_start=5, width=800, heigth=600 )
```

```python
# Recuperando a quantidade de pontos
qtde = len(df_convertido)
```

```python
qtde
```

```python
for i in range(qtde):
    lng = float(df_convertido.iloc[i]['Longitude'])
    lat = float(df_convertido.iloc[i]['Latitude'])
    desc = 'Ponto Turistico: ' + str(df_convertido.iloc[i]['Ponto Turístico']) + '<br> Cidade: ' +  str(df_convertido.iloc[i]['Cidade']) +'<br> Latitude: ' +  str(df_convertido.iloc[i]['Latitude']) + '<br> Longitude: ' +  str(df_convertido.iloc[i]['Longitude']) + '<br> : Média Anual de Visitação: ' +  str(df_convertido.iloc[i]['Média Anual de Visitação'])
    folium.Marker ([lat, lng], popup = desc, icon = folium.Icon(color='red')).add_to(locationmap)
```
